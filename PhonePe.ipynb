```python
# Merge both district and state-level data
merged_summary = district_summary.merge(state_summary, on='State', suffixes=(' District', ' State'))
```

```python
# Create match checks for each metric
merged_summary['transactions_match'] = merged_summary['Transactions District'] == merged_summary['Transactions State']
merged_summary['amount_match'] = merged_summary['Amount (INR) District'] == merged_summary['Amount (INR) State']
merged_summary['users_match'] = merged_summary['Registered Users District'] == merged_summary['Registered Users State']
```

```python
# Show mismatches
discrepancies_summary = merged_summary[~(merged_summary['transactions_match'] & merged_summary['amount_match'] & merged_summary['users_match'])]
```

```python
# Print the discrepancies if there are any
print(discrepancies_summary[['State', 'Transactions District', 'Transactions State', 'Amount (INR) District', 
                             'Amount (INR) State', 'Registered Users District', 'Registered Users State']])
```
